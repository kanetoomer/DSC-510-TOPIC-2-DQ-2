```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

I am creating skewed income data

```python
np.random.seed(0)
low_income = np.random.exponential(scale=20000, size=1000)
high_income = np.random.normal(loc=20000, scale=50000, size=10)
income_data = np.concatenate([low_income, high_income])
```

This is my dataframe

```python
df = pd.DataFrame(income_data, columns=['Income'])
df = df[df['Income'] > 0]
```

Getting the mean and median

```python
mean_income = df['Income'].mean()
median_income = df['Income'].median()
```

Log transformation

```python
df['Log_Income'] = np.log1p(df['Income'])
```

My visuals for the original and log transformed data

```python
fig, axes = plt.subplots(nrows=1, ncols=2, figsize=(14, 6))
df['Income'].hist(bins=30, ax=axes[0], color='skyblue', edgecolor='black')
axes[0].set_title('Original Income Data')
axes[0].set_xlabel('Income')
axes[0].set_ylabel('Frequency')

df['Log_Income'].hist(bins=30, ax=axes[1], color='lightgreen', edgecolor='black')
axes[1].set_title('Log-Transformed Income Data')
axes[1].set_xlabel('Log(Income)')
axes[1].set_ylabel('Frequency')

plt.tight_layout()
plt.show()
```

Printing my mean and median values

```python
print(f"Original Mean: {mean_income}, Original Median: {median_income}")
print(f"Log-Transformed Mean: {df['Log_Income'].mean()}, Log-Transformed Median: {df['Log_Income'].median()}")
```
